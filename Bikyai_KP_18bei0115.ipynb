```python
from urllib.request import urlopen
import json

def json_fetcher(url='http://api.nobelprize.org/v1/prize.json'):
  response = urlopen(url)
  data_json = json.loads(response.read())
  out_file = open("data.json", "w") 
  json.dump(data_json, out_file, indent = 6) 
  out_file.close()
  return data_json 

#Function call
data_json=json_fetcher()
```

```python
import pandas as pd
count=0

output = pd.DataFrame()
for (k, v) in data_json.items():
  for values in v:
    temp_dict = values
    #print(temp_dict['category'])
    try:
        #print(temp_dict['laureates'])
        for dicts in temp_dict['laureates']:
          dicts.update({'category':temp_dict['category'],'year':temp_dict['year']})
          print(dicts)
          output=output.append(dicts,ignore_index=True)
          
    except:
        print("field does not exist")
    
```
